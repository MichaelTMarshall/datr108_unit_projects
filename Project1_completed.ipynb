```python
#Challenge 1: Largest Palindrome

num = 0                            #I have assigned the variable num the value of 0
for a in range(1000, 99, -1):      #I have created a for loop ranging from 999 counting downwards by 1 to 100
    for b in range(a, 99, -1):     #Within that for loop I have created another for loop to count from the current number to 100 down by 1
        total = a * b              #creates the variable total and assigns the value of total to be the product of each cuurent number in the for loops
        if total > num:            #if statement declaring if the variable total is greater then the variable num to proceed
            c = str(total)         #creates variable c and assigns the value of total in string form
            if c == c[::-1]:       #if statement indicating if the value of c equals the value of c backwards then to continue
                num = total        #reassigns the value of variable num to the the value of total
print(num)                         #prints the new value of the variable num
```

```python
#Challenge 2: Summation of Primes

def isPrime(n):                                 #creates the function isPrime and allows input to be defined by n
    if n < 2: return "not prime, too small!"    #if statement stating if 2 is less then 2 return "not prime, too small!"
    for i in range(2, int(n*0.5) + 1):          #for loop counting from that calculates if a number is prime by returning only prime values starting at 3
        if n % i == 0:
            return False
    return True

sum = 0                                         #assigns the variable sum the value of 0
for i in range(2, 2000):                        #for loop that counts from 2 to 1999
    if isPrime(i):                              #if statement stating if the current number in the for loop is a number that was passed from the isPrime function, to add to the sum of all prime numbers
        sum += i

print (sum)                                     #prints the sum value, which is the total sum of all numbers deemed prime from 0 to 1999
```

```python
#Challenge 3: Multiples of 3 and 5

count = 0                           #assigns the variable count the value 0
for n in range(1 , 1000):           #for loop that counts from 1 to 999
    if n % 3 == 0 or n % 5 == 0:    #if statement declaring if the number in for loop is divisible by either 3 or 5 then to add that number to the variable count  
        count += n
    print(count)                    #prints the total of variable count, the total of all numbers divisible by 3 or 5 below 1000
```

```python
#Challenge 4: String Compressor

def compress(string):                                 #creates funciton compress assigning the variable string to the functions input
    a = ""                                            #assigns variable a the value of a string with no value
    count = 1                                         #assigns variable count to the integer 1
    a += string[0]                                    #adds the size of the imput variable string to the variable a starting at the first letter in the string
                                                      
    for i in range(len(string)-1):                    #for loop that ranges the length of the variable string, skipping the last one
        if(string[i] == string[i+1]):                 #if statement if the value of i equals the value of i + 1 then to raise the count, this means that if the letter in the string is equal to the letter directly behind it then to raise the count
            count+=1
        else:                                         #else statement for when the value of i does not match the value of i + 1
            if(count > 1):                            #if statement witihin the else statement that if the count is greater then 1 then add the value of the string plus the count in string form following the string value
                #Ignore if no repeats
                a += str(count)
            a += string[i+1]                          #this moves to the next letter in the string
            count = 1                                 #this resets the count to then rerun the procedure for the next letter
    #print last one
    if(count > 1):                                    #if statement for the last letter in the string stating if the letter is followed by a similar letter to print the letter followed by the count, if not to print the letter of the string
        a += str(count)
    return a                                          #returns the full compressed string, once the function is given an input

                                                      #my function does not account for case sensitivity but could convert if adding the .lower() or .upper() after the strings input into the function
```
